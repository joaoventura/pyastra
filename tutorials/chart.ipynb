Do the necessary imports, create datetime and position objects and build the chart.

```python
from pyastra import const
from pyastra import Chart
from pyastra import Datetime
from pyastra import GeoPos


# Build a chart for a date and location
date = Datetime('2015/03/13', '17:00', '+00:00')
pos = GeoPos('38n32', '8w54')
chart = Chart(date, pos)

# Print chart
print(chart)
```

Planets and house cusps can also be accessed from the chart object.

```python
# Print objects and houses
print(chart.get(const.SUN))
print(chart.get(const.ASC))
```
